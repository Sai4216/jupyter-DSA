```python
def split_casewise(s):
    if not s:
        return []
    result = []
    new_word = s[0]
    for char in s[1:]:
        if (char.isupper() and new_word[-1].islower()) or (char.islower() and new_word[-1].isupper()):
            result.append(new_word)
            new_word = char
        else:
            new_word += char
    result.append(new_word)
    return result

s = "ABcdEFgh"
print(split_casewise(s))
# Output: ['AB', 'cd', 'EF', 'gh']
```

```python
def f(arr):
    n = len(arr)
    res = []
    res.append(arr[0])
    for i in range(1,n):
        if arr[i] not in res:
            res.append(arr[i])
        else:
            if arr[i]+1 not in res:
                res.append(arr[i]+1)
            elif arr[i]-1 not in res and arr[i] != 1:
                res.append(arr[i]-1)
    return len(res)
f([3,3,5,5,3])
```

```python
def f(capacities):
    capacities.sort()
    used = set()
    
    for cap in capacities:
        # Try to use cap - 1, if it's possible and > 0 and not already used
        if cap - 1 > 0 and (cap - 1) not in used:
            used.add(cap - 1)
        # Otherwise use cap if not used
        elif cap not in used:
            used.add(cap)
        # Otherwise use cap + 1 if not used
        elif (cap + 1) not in used:
            used.add(cap + 1)
        # If none is available, skip (not possible to use more)
    print(used)
    return len(used)

f([3,3,5,5,3])
```

```python
edges = [(0,1),(1,2),(1,5),(1,7),(2,3),(2,5),(2,6),(3,5),(3,4)]
n_nodes = 8
```

```python
class Graph:
    def __init__(self,edges,n_nodes):
        self.data = [[] for _ in range(n_nodes)] 
        for n1,n2 in edges:
            self.data[n1].append(n2)
            self.data[n2].append(n1)
    def __repr__(self):
        return "\n".join(["{} : {}".format(node,edge) for node,edge in enumerate(self.data)])
    
```

```python
graph1 = Graph(edges,n_nodes)
```

```python
def bfs(graph,root):
    visited = [False]*len(graph.data)
    queue = []
    queue.append(root)
    visited[root] = True
    idx = 0
    while idx < len(queue):
        current = queue[idx]
        for node in graph.data[current]:
            if not visited[node]:
                queue.append(node)
                visited[node] = True
        idx += 1
    return queue
```
