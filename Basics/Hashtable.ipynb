```python
list = [None] * max_size
```

```python
class HashTable:
    def __init__(self , max_size = max_size):
        self.list = [None] * max_size
    def get_index(self  , key ):
        sum = 0
        for character in key:
            number = ord(character)
            sum += number
        indx = sum % len(list)
        return indx
    def insert(self , key , value):
        indx = self.get_index(key)
        print(indx)
        self.list[indx] = key,value
    def find(self , key):
        indx = self.get_index(key)
        kv = self.list[indx]
        if kv is None:
            return 'value is not found please insert it first'
        else:
            key,value = kv
            return value
    def update(self , key ,value):
        idx = self.get_index(key)
        self.list[idx] = key , value
        # kv = self.list[idx]
        # if kv is not None:
        #     kv[1] = value 
    def list_all(self):
        return [kv[0] for kv in self.list if kv is not None]
    
```
