```python
test1 = {
    'input' : {'list' : [2,3,4,5,6,7,8,9],'query' : 2},
    'output' : 0
}
test2 = {
    'input' : {'list' : [2,3,4,5,6,7,8,9],'query' : 55},
    'output' : -1
}
test3 = {
    'input' : {'list' : [],'query' : 3},
    'output' : -1
}
test4 = {
    'input' : {'list' : [2,3,4,5,6,7,8,9],'query' : 9},
    'output' : 7
}
```

```python
tests = [test1 , test2 ,test3 , test4 ]
```

```python
%%time
def Binary_search(list , query):
    low = 0
    high = len(list) - 1
    while low <= high:
        mid = (low + high) // 2
        if list[mid] < query:
            low = mid + 1
        elif list[mid] > query:
            high = mid - 1
        else:
            return mid
    return -1
```

```python
Binary_search(**test2['input'] )== test2['output']
```

```python
for test in tests:
    print(test)
    print(Binary_search(**test['input']) == test['output'])
```

```python
class user:
    def __init__(self , username , useremail):
        self.username = username
        self.useremail = useremail
    def __repr__(self):
        return 'User(username = {} , useremail = {})'.format(self.username,self.useremail)
    def __str__(self):
        return self.__repr__()
```

```python
class database:
    def __init__(self):
        self.users = []
    def insert(self , user):
        i = 0
        while i < len(self.users):
            if user.username < self.users[i].username:
                break
            i += 1
        self.users.insert(i,user)
    def find(self , username):
        for target in self.users:
            if username == target.username:
                return target
            else:
                return "{} is not found in user's list".format(username)
    def update(self,user):
        target = self.find(user.username)
        target.useremail = user.useremail
        
    def listall(self):
        return self.users
```

```python
database = database()
```

```python
def tuple_to_tree(data):
    if isinstance(data , tuple) and len(data) == 3:
        node = TreeNode(data[1])
        node.left = tuple_to_tree(data[0])
        node.right = tuple_to_tree(data[2])
    elif data == None:
        node = None
    else:
        node = TreeNode(data)
    return node
```
