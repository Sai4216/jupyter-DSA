```python
#Binary_search
def binary_search(arr,target,low,high):
    mid = (low + high) // 2
    if low > high:
        return False
    if arr[mid] == target:
        return True
    elif target < arr[mid]:
        high = mid - 1
    else:
        low = mid + 1
    return binary_search(arr,target,low,high)

arr = [1, 2, 3, 5, 7, 9, 12]
target = 4
n = len(arr)
print(binary_search(arr,target,0,n-1))
```

```python
#Interpolation search
def interpolation_search(arr): #For uniformly distributed array
    low = 0
    high = len(arr) - 1
    while low < high and target >= arr[low] and target <= arr[high]:
        #For one len arr
        if low == high:
            if arr[low] == target:
                return True
            return False
        pos = low + int(((target-arr[low]) / (arr[high]-arr[low])) * (high-low))
        if target == arr[pos]:
            return True
        elif target < arr[pos]:
            high = pos - 1
        else:
            low = pos + 1
    return False

arr = [10,20,30,40,50,60,70,80,90]
target = 80
print(interpolation_search(arr))
```

```python
def ternary_search(arr,target):
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid1 = low + (high-low) // 3
        mid2 = high - (high-low) // 3
        if arr[mid1] == target:
            return mid1
        if arr[mid2] == target:
            return mid2
        if target < arr[mid1]:
            high = mid1-1
        elif target > arr[mid2]:
            low = mid2+1
        else:
            low = mid1+1
            high = mid2-1
    return -1

arr = [1,2,3,4,5,6,12,16,20]
target = 22
ternary_search(arr,target)
```

```python
#exponential Search
def exponential_search(arr,target):
    n = len(arr)
    if arr[0] == target:
        return 0
    i = 1
    while i<n and arr[i] <= target:
        i = i*2
    return binary_search(arr,target)

arr = [1,2,3,4,5,6,12,16,20]
target = 6
```

```python
#Ternary Search
def ternary_search(arr,low,high):
    if low == high:
        return low
    mid1 = low + (high-low)//3
    mid2 = high - (high-low)//3
    if arr[mid1] < arr[mid2]:
        low = mid1+1
    else:
        high = mid2-1
    return ternary_search(arr,low,high)

    
arr = [1,5,6,28,23,20,15,7]
low = 0
high = len(arr)-1
ternary_search(arr,low,high)
```
