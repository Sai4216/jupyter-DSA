```python
def f(g,s):
    l,r = 0,0
    n , m = len(g) , len(s)
    s.sort()
    g.sort()
    while l<n and r<m:
        if g[l] <= s[r]:
            l += 1
        
        r += 1
    return l
print(f([10,9,8,7],[5,6,7,8]))
```

```python
# knapsack
def f(val,wt,c):
    ans = 0
    i = 0
    while c > 0 and i < len(wt):
        if c >= wt[i]:
            c -= wt[i]
            ans += val[i]
        else:
            ans += (val[i]/wt[i])*c
        i += 1
    return ans
    
val = sorted(list(map(int,input().split())))
wt = sorted(list(map(int,input().split())))
capacity = int(input())
ans = f(val,wt,capacity)
print("%.6f"%ans)
```

```python
#customer bill
def f(bills):
    five = 0
    ten = 0
    for i in bills:
        if i == 5:
            five += 1
        elif i == 10:
            if five:
                five -= 1
                ten += 1
            else:
                return False
        else:
            if five and ten:
                five -= 1
                ten -= 1
            elif five >= 3:
                five -= 3
            else:
                return False
    return True
    
bills = [5,5,5,10,20]
f(bills)
```

```python
#valid paranthsis

def f(s,ind,cnt):
    if ind < 0:
        return False
    if ind == len(s):
        return cnt == 0
    if s[ind] == "(":
        return f(s,ind+1,cnt+1)
    elif s[ind] == ')':
        return f(s,ind+1,cnt-1)
    else:
        return f(s,ind+1,cnt+1) or f(s,ind+1,cnt-1) or f(s,ind+1,cnt)
    
s = "(((((*(()((((*((**(((()()*)()()()*((((**)())*)*)))))))(())(()))())((*()()(((()((()*(())*(()**)()(())"
f(s,0,0)
```

```python
# Jump Game to reach end of the array
def f(arr,n):
    maxInd = 0
    for i in range(n):
        if i > maxInd:
            return False
        maxInd = max(maxInd,arr[i] + i)
        # print(i,maxInd)
    return True

arr = [3,2,1,0,4]
n = len(arr)
f(arr,n)
```

```python
# 57. Insert Interval
def f(intervals, newInterval):
    ans = []
    n = len(intervals)
    i = 0
    while i<n and intervals[i][1] < newInterval[0]:
        ans.append(intervals[i])
        i += 1
    while i<n and intervals[i][0] < newInterval[1]:
        newInterval[0] = min(newInterval[0],intervals[i][0])
        newInterval[1] = max(newInterval[1],intervals[i][1])
        i += 1
    ans.append(newInterval)
    while i<n:
        ans.append(intervals[i])
        i += 1
    return ans


intervals = [[1,2],[3,5],[6,7],[8,10],[12,16]]
newInterval = [4,8]
f(intervals,newInterval)
```
