```python
#fibanocci
def f(n,dp):
    if n <= 1:
        return n
    if dp[n] != -1:
        return dp[n]
    dp[n] = f(n-1,dp) + f(n-2,dp)
    return dp[n]

n = int(input())
dp = [-1] * (n+1)
f(n,dp)
```

```python
#L51
def safe(row,col,board,n):
    dupcol = col
    duprow = row
    if col == 0 and row == 0:
        return True
    while col >= 0 and row >= 0:
        if board[row][col] == 'Q': return False
        col -= 1
        row -= 1
    col = dupcol
    row = duprow
    while col >= 0:
        if board[row][col] == 'Q': return False
        col -= 1
    col = dupcol
    row = duprow
    while col >= 0 and row < n:
        if board[row][col] == 'Q': return False
        col -= 1
        row += 1
    return True

def f(col,board,ans,n):
    if col >= n:
        ans.append(list(board))
        return
    for row in range(n):
        if safe(row,col,board,n):
            board[row] = board[row][:col] + 'Q' + board[row][col+1:]
            f(col+1,board,ans,n)
            board[row] = board[row][:col] + '.' + board[row][col+1:]
    return ans

n = 4
ans = []
board = ['.'*n for i in range(n)]
f(0,board,ans,n)
```
