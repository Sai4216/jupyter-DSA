```python
import sys
def f(height,ind,dp):
    if ind <= 0: return 0
    if dp[ind] != -1:
        return dp[ind]
    right = sys.maxsize
    left = f(height,ind-1,dp) + abs(height[ind]-height[ind-1])
    if ind > 1:
        right = f(height,ind-2,dp) + abs(height[ind]-height[ind-2])
    dp[ind] = min(left,right)
    return dp[ind]

height = [2, 1, 3, 5, 4]
dp = [-1] * len(height)
f(height,len(height)-1,dp)
```

```python
def f(ind,arr,n,dp):
    if ind >= n:
        return 0
    if dp[ind] != -1:
        return dp[ind]
    pick = arr[ind] + f(ind+2,arr,n,dp)
    nonpick = 0 + f(ind+1,arr,n,dp)
    dp[ind] = max(pick,nonpick)
    return dp[ind]

arr = [2,7,9,3,1]
n = len(arr)
dp = [-1]*n
f(0,arr,n,dp)
```
