```python
def frequencyCount(a,b):
    union = []
    i, j = 0, 0
    while i<len(a) and j < len(b):
        if a[i] <= b[j]:
            if len(union) == 0 or a[i] != union[-1]:
                union.append(a[i])
            i += 1
        else:
            if len(union) == 0 or b[j] != union[-1]:
                union.append(b[j])
            j += 1
    while i < len(a):
        if union[-1] != a[i]:
            union.append(a[i])
        i += 1
    while j < len(b):
        if union[-1] != b[j]:
            union.append(b[j])
        j += 1
    return union
print(frequencyCount([-5,-4,-1,1,7],[-3,1,8]))
```

```python
class Solution(object):

    def helper(self,i,arr,n,dp):
        if i >= n:
            return 0
        if dp[i] != -1:
            return dp[i]
        
        pick = arr[i] + self.helper(i+2,arr,n,dp)
        nonpick = 0 + self.helper(i+1,arr,n,dp)
        dp[i] = max(pick,nonpick)
        return dp[i]


    def rob(self, nums):
        n = len(nums)
        arr1 = nums[:n-1]
        arr2 = nums[1:]
        dp1 = [-1] * len(arr1)
        dp2 = [-1] * len(arr2)
        return max(self.helper(0,arr1,len(arr1),dp1),self.helper(0,arr2,len(arr2),dp2))
ob = Solution()
ob.rob([2,3,2])
```

```python

# Function to solve the problem using dynamic programming
def solveUtil(ind, arr, dp):
    # Check if the solution for this index has already been calculated
    if dp[ind] != -1:
        return dp[ind]
    
    # Base case: when the index is 0, return the value at that index
    if ind == 0:
        return arr[ind]
    
    # Base case: when the index is negative, return 0 (out of bounds)
    if ind < 0:
        return 0
    
    # Calculate the maximum value when picking the current element
    pick = arr[ind] + solveUtil(ind - 2, arr, dp)
    
    # Calculate the maximum value when not picking the current element
    nonPick = 0 + solveUtil(ind - 1, arr, dp)
    
    # Store the maximum of the two choices in the DP table
    dp[ind] = max(pick, nonPick)
    
    # Return the maximum value for the current index
    return dp[ind]

# Function to solve the problem for the given array
def solve(n, arr):
    # Initialize a DP table with -1 values to store intermediate results
    dp = [-1 for i in range(n)]
    
    # Call the recursive utility function to find the maximum value
    return solveUtil(n - 1, arr, dp)

# Main function to test the code
def main():
    arr = [1,3,2,1]
    n = len(arr)
    
    # Call the solve function and print the result
    print(solve(n, arr))

if __name__ == '__main__':
    main()

```

```python
def solve(day,last,arr,n,dp):
    if dp[day][last] != -1:
        return dp[day][last]
    if day == n-1:
        maxi = 0
        for i in range(3):
            if i != last:
                maxi = max(maxi,arr[day][i])
        return maxi
    maxi = 0
    for i in range(3):
        if i != last:
            points = arr[day][i] + solve(day+1,i,arr,n,dp)
            maxi = max(maxi,points)
    dp[day][last] = maxi
    return dp[day][last]

arr = [[1, 2, 5], [3, 1, 1], [3, 3, 3]]
n = len(arr)
dp = [[-1 for i in range(4)] for j in range(n)]
solve(0,3,arr,n,dp)
```

```python
# printing permutations
# L46
from collections import defaultdict
def f(ds,arr,mp,ans):
    if len(ds) == len(arr):
        ans.append(list(ds))
        return
    for i in range(len(arr)):
        if mp[i] != 1:
            mp[i] = 1
            ds.append(arr[i])
            f(ds,arr,mp,ans)
            ds.pop()
            mp[i] = 0
    

arr = [1,2,3]
mp = defaultdict(int)
ans = []
f([],arr,mp,ans)
ans
```
