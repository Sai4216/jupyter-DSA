```python
def f(nums):
    pos = []
    neg = []
    for i in range(len(nums)):
        if nums[i] > 0:
            pos.append(nums[i])
        else:
            neg.append(nums[i])
    for i in range(len(pos)):
        nums[i*2] = pos[i]
    for i in range(len(neg)):
        nums[i*2+1] = neg[i]
    return nums
    
            

nums = [1,-8,-9,-5,5,4]
print(f(nums))
```

```python
from collections import defaultdict
mpp = defaultdict(int)
arr = [1,1,1]
k = 2
mpp[0] = 1
preSum = 0
cnt = 0
for i in range(len(arr)):
    preSum += arr[i]
    remove = preSum - k
    cnt += mpp[remove]
    mpp[preSum] += 1
print(cnt)
```

```python
# leaders of the array
import sys
def f(arr):
    n = len(arr)
    ans = []
    maxi = -sys.maxsize
    for i in range(n-1,-1,-1):
        if arr[i] > maxi:
            ans.append(arr[i])
            maxi = arr[i]
    ans.reverse()
    return ans
arr = [16, 17, 4, 3, 5, 2]
print(f(arr))
```

```python
#next permutation

def f(ind,arr):
    n = len(arr)
    for i in range(n-2,-1,-1):
        if arr[i] < arr[i+1]:
            ind = i
            break
    if ind == -1:
        arr.reverse()
        return arr
    for i in range(n-1,ind,-1):
        if arr[i] > arr[ind]:
            arr[i], arr[ind] = arr[ind],arr[i]
            break
    arr[ind+1:] = reversed(arr[ind+1:])
    return arr


arr = [3,2,1]
f(-1,arr)
```

```python
#maximum product of the array
import sys
def f(arr):
    n = len(arr)
    maxi = float('-inf')
    pref = 1
    suff = 1
    for i in range(n):
        if pref == 0: pref = 1
        if suff == 0: suff == 1
        pref *= arr[i]
        suff *= arr[n-i-1]
        maxi = max(maxi,max(pref,suff))
    return maxi
f([2,3,-2,4])
```
