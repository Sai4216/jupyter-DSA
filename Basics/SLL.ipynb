```python
class Node:
    def __init__(self,val):
        self.val = val
        self.next = None
```

```python
def covertArr(arr):
    head = Node(arr[0])
    temp = head
    for i in range(1,len(arr)):
        n = Node(arr[i])
        temp.next = n
        temp = n
    return head
def find(head,key):
    temp = head
    while temp:
        if temp.val == key:
            print('found')
            return
        temp = temp.next
    print('not found')
def delete(head,key):
    temp = head
    prev = head
    while temp:
        if temp.val == key:
            prev.next = temp.next
        prev = temp
        temp = temp.next
def Print(head):
    temp = head
    while temp:
        print(temp.val)
        temp = temp.next
```

```python
arr = [1,2,3,4,5,6]
head = covertArr(arr)
```
