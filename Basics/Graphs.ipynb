```python
num_nodes = 5
edges = [(0,1),(0,4),(1,2),(1,3),(1,4),(2,3),(3,4)]
num_nodes , len(edges)
```

```python
class Graph:
    def __init__(self , num_nodes , edges):
        # self.dic = {}
        self.num_nodes = num_nodes
        self.data = [[] for _ in range(num_nodes)]
        for n1,n2 in edges:
            self.data[n1].append(n2)
            self.data[n2].append(n1)
        # for ind , val in enumerate(self.data):
        #     self.dic[ind] = val
    def addEdge(self , edge):
        for n1,n2 in edge:
            self.data[n1].append(n2)
            self.data[n2].append(n1)
            return 'Edge is added'
    def removeEdge(self ,edge):
        for n1 , n2 in edge:
            if n1 in self.data[n2] and n2 in self.data[n1]:
                self.data[n2].remove(n1)
                self.data[n1].remove(n2)
                return 'Edge is removed'
    def __repr__(self):
        return '\n'.join(['{} : {}'.format(ind , val) for ind , val in enumerate(self.data)])
    # def __str__(self):
    #     return self.__repr__()
```

```python
graph1 = Graph(num_nodes , edges)
```

```python

def bfs(graph,root):
    visited = [False]*len(graph.data)
    queue = []
    visited[root] = True
    queue.append(root)
    idx = 0
    while idx < len(queue):
        current = queue[idx]
        for node in graph.data[current]:
            if not visited[node]:
                queue.append(node)
                visited[node] = True
        idx += 1
    return queue
```

```python
bfs(graph1,4)
```

```python
visited = []
def dfs(graph,node):
    if node in visited:
        return
    print(node,end=" ")
    visited.append(node)
    for neighbor in graph.data[node]:
        if neighbor not in visited:
            dfs(graph,neighbor)
    return visited
```
