```python
## hash Map
class HashTable():
    def __init__(self):
        self.max = 100
        self.arr = [None] * self.max
    def gethash(self,key):
        idx = 0
        for i in key:
            idx += ord(i)
        h = idx % self.max
        return h
    def __setitem__(self,key,val):
        h = self.gethash(key)
        self.arr[h] = val
    def __getitem__(self,key):
        h = self.gethash(key)
        return self.arr[h]
    def __delitem__(self,key):
        h = self.gethash(key)
        self.arr[h] = None
```

```python
t = HashTable()

t.gethash('sai')

t['sai'] = 10
t.arr

t['sai']

t.arr
```

```python
##Stack using list
class Stack():
    def __init__(self):
        self.arr = []
    def push(self,x):
        self.arr.append(x)
    def pop(self,ind):
        return self.arr.pop(ind)
    def peek(self):
        return self.arr[-1]
    def size(self):
        return size(self.arr)
```

```python
s = Stack()
s.push(5)
s.push(50)
s.push(1)
s.pop(1)
```

```python
#tree
class TreeNode():
    def __init__(self,str):
        self.data = str
        self.children = []
        self.parent = None
    def add_child(self,child):
        self.children.append(child)
        child.parent = self
    def get_level(self):
        level = 0
        p = self.parent
        while p:
            level += 1
            p = p.parent
        return level
    def print_tree(self):
        spaces = " " * self.get_level() * 3
        pre = "|--" if self.parent else ""
        print(spaces + pre + self.data)
        for child in self.children:
            child.print_tree()


root = TreeNode('electronics')
mobile = TreeNode('mobile')
tv = TreeNode('tv')
laptop = TreeNode('laptop')
mobile.add_child(TreeNode('apple'))
mobile.add_child(TreeNode('vivo'))
mobile.add_child(TreeNode('realme'))
tv.add_child(TreeNode('LG'))
tv.add_child(TreeNode('Skyworth'))
tv.add_child(TreeNode('SAMSUNG'))
laptop.add_child(TreeNode('hp'))
laptop.add_child(TreeNode('Dell'))
laptop.add_child(TreeNode('Lenovo'))
root.add_child(mobile)
root.add_child(tv)
root.add_child(laptop)
```

```python
#BinarySearchTree
class BSTNode():
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
    def add_to_tree(self,data):
        if self.data == data:
            return
        elif data < self.data:
            if self.left:
                self.left.add_to_tree(data)
            else:
                self.left = BSTNode(data)
        else:
            if self.right:
                self.right.add_to_tree(data)
            else:
                self.right = BSTNode(data)
    def inorder(self):
        elements = []
        if self.left:
            elements += self.left.inorder()
        elements.append(self.data)
        if self.right:
            elements += self.right.inorder()
        return elements

node_values = [25,5,47,3,94,9,8,6,16,58,45]
root = BSTNode(node_values[0])
for i in range(1,len(node_values)):
    root.add_to_tree(node_values[i])
```
