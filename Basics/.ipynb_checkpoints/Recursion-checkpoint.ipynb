```python
def f(cnt,n):
    if cnt > n:
        return
    print(cnt)
    f(cnt+1,n)

n = int(input())
f(1,n)
```

```python
#Back-Tracking
def f(i,n):
    if i > n:
        return
    f(i+1,n)
    print(i)
n = int(input())
f(1,n)
```

```python
#factorial
def f(n):
    if n == 1:
        return 1
    return n * f(n-1)
    
n = int(input())
f(n)
```

```python
#reverse
def f(i,n):
    if i >= n//2:
        return
    arr[i], arr[n-i-1] = arr[n-i-1], arr[i]
    f(i+1,n)

arr = list(map(int,input().split()))
n = len(arr)
f(0,n)
arr
```

```python
#pallindrom
def f(i,n):
    if i >= n // 2:
        return True
    if s[i] != s[n-i-1]:
        return False
    f(i+1,n)
    
s = input()
n = len(s)
print(f(0,n))
```

```python
#fibanocci
def f(n,dp):
    if n <= 1:
        return n
    if dp[n] != -1:
        return dp[n]
    return f(n-1,dp) + f(n-2,dp)

n = int(input())
dp = [-1]* (n+1)
f(n,dp)
```
