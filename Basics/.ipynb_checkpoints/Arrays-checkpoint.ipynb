```python
def find_position(array,number):
    i = 0
    for i in range(len(array)):
        if array[i] == number:
            return i
    return -1
```

```python
test1 = {
    'input' : {'array' : [],'number' : 4},
    'output' : -9
}
```

```python
test1['output']==find_position(**test1['input'])
```

```python

def maxFrequencyElements( nums):
    """
    :type nums: List[int]
    :rtype: int
    """
    list_nums = list(nums)
    dic ={}
    for i in list_nums:
        if i not in dic:
            dic[i] = 1
        else:
            dic[i] += 1
    print(dic)
    maximum = 0
    l = []
    for i in dic:
        l.append(dic[i])
        maximum = max(maximum , dic[i])
    print(l)
    return l.count(maximum)*maximum



maxFrequencyElements([1,2,2,3,1,4])
```

```python

def groupAnagrams( strs):
    """
    :type strs: List[str]
    :rtype: List[List[str]]
    """
    dic ={}
    i = 0
    for string in strs:
        hash = 0
        for j in string:
            hash += ord(j)
        # print(hash,string)
        if hash not in dic:
            dic[hash] = list()
        # print(dic[hash])
        l = dic[hash]
        l.append(string)
        # dic[hash] = dic[hash].append(string)
        # i += 1
        # print(dic)
    li = []
    for i in dic:
        li.append(dic[i])
    return li
```

```python
print(groupAnagrams(["eat","tea","tan","ate","nat","bat"]))
```
