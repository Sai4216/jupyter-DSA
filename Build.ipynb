```python
class Node:
    def __init__(self,val):
        self.val = val
        self.next = None
def covertArr(arr):
    head = Node(arr[0])
    temp = head
    for i in range(1,len(arr)):
        n = Node(arr[i])
        temp.next = n
        temp = n
    return head
```

```python
arr = [1]
head = covertArr(arr)
temp = head
while temp:
    pre = temp
    print(temp.val)
    temp = temp.next
```

```python
n = 1
n = n+1
first = head
if not first or not first.next:
    print(None)
while n>0:
    first = first.next
    n = n-1
second = head
while first:
    first = first.next
    second = second.next
if second.next:
    second.val = second.next.val
    second.next = second.next.next
    temp = head
    while temp:
        pre = temp
        print(temp.val)
        temp = temp.next
else:
    print(None)
```
