```python
class Node:
    def __init__(self,val):
        self.val = val
        self.next = None
def covertArr(arr):
    head = Node(arr[0])
    temp = head
    for i in range(1,len(arr)):
        n = Node(arr[i])
        temp.next = n
        temp = n
    return head
```

```python
from collections import defaultdict
mp = defaultdict(Node)
arr = [2,4,3]
head = covertArr(arr)
temp = head
while temp:
    # print(temp.val)
    mp[temp]= 1
    temp = temp.next
```
