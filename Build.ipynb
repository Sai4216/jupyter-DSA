```python
class Solution(object):
    def helper(self,i,target,arr,n,dp):
        if i >= n:
            return False
        if target == 0:
            return True
        if dp[i][target] != -1:
            return dp[i][target]
        notTake = self.helper(i+1,target,arr,n,dp)
        take = False
        if arr[i] <= target:
            take = self.helper(i+1,target-arr[i],arr,n,dp)
        dp[i][target] = take or notTake
        return dp[i][target]

    def canPartition(self, nums):
        n = len(nums)
        total = sum(nums)
        
        if total % 2 == 0:
            target = total // 2
            dp = [[-1 for i in range(target + 1)] for j in range(n)]
            print(dp)
            return self.helper(0,target,nums,n,dp)
        return False

n = Solution().canPartition([1,5,11,5])
n
```
