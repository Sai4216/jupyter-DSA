```python
def f(i,arr,n,target,ds,ans):
    if i >= n:
        if target == 0:
            ans.append(list(ds))
        return
    ds.append(arr[i])
    f(i+1,arr,n,target-arr[i],ds,ans)
    ds.pop()
    f(i+1,arr,n,target,ds,ans)
    return ans

arr = [10,1,2,7,6,1,5]
target = 8
f(0,arr,len(arr),target,[],[])
```
