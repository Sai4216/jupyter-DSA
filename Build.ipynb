```python
def f(ind,arr,n,target,ds,ans):
    if target == 0:
        ans.append(list(ds))
    return
    print(ds)
    for i in range(ind,n):
        if i>ind and ar[i] == arr[i-1]: continue
        if arr[i] > target: break
        ds.append(arr[i])
        f(i+1,arr,n,target-arr[i],ds,ans)
        ds.pop()
    return ans

arr = sorted([2,5,2,1,2])
# arr = [2,5,2,1,2]
print(arr)
target = 5
f(0,arr,len(arr),target,[],[])
# print("ji")
# print(ans)
```

```python
def f(ind,n,k,ans,ds):
    if n == 0 and len(ds) == k:
        ans.append(list(ds))
        return
    if len(ds)>k or n<0:return
    for i in range(ind,10):
        if i>n:return
        ds.append(i)
        f(i+1,n-i,k,ans,ds)
        ds.pop()
    return ans

n = 7
k = 3
ans = []
f(1,n,k,ans,[])
print(ans)
```
