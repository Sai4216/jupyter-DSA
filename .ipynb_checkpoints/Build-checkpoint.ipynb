```python
class Node:
    def __init__(self,val):
        self.val = val
        self.next = None
def covertArr(arr):
    head = Node(arr[0])
    temp = head
    for i in range(1,len(arr)):
        n = Node(arr[i])
        temp.next = n
        temp = n
    return head
```

```python
arr = [3,2,0,-4]
head = covertArr(arr)
# temp = head
```

```python
temp = head
while temp.next:
    pre = temp
    print(temp.val)
    temp = temp.next
# print(pre.val)
temp.next = pre
```

```python
def hasCycle(head):
    temp = head
    mp = {}
    # c = 0
    while temp:
        if temp.val in mp:
            return mp[temp.val]
        mp[temp.val] = temp
        # c += 1
        temp = temp.next
    return -1
node = hasCycle(head)
node.val
```
