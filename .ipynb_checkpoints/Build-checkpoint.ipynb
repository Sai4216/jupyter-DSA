```python
class Node:
    def __init__(self,val):
        self.val = val
        self.next = None
def covertArr(arr):
    head = Node(arr[0])
    temp = head
    for i in range(1,len(arr)):
        n = Node(arr[i])
        temp.next = n
        temp = n
    return head
```

```python
arr = [1,2,1,5,7]
head = covertArr(arr)
temp = head
while temp:
    pre = temp
    print(temp.val)
    temp = temp.next
```
