```python
def f(ind,arr,n,target,ds,ans):
    if target == 0:
        ans.append(list(ds))
    return
    for i in range(ind,n):
        if i>ind and ar[i] == arr[i-1]: continue
        if arr[i] > target: break
        ds.append(arr[i])
        f(i+1,arr,n,target-arr[i],ds,ans)
        ds.pop()
    return ans

arr = sorted([2,5,2,1,2])
# arr = [2,5,2,1,2]
print(arr)
target = 5
f(0,arr,len(arr),target,[],[])
```
