```python
def is_bst(node):
    if node is None:
        return True,None,None
    is_bst_l , l_min , l_max = is_bst(node.left)
    is_bst_r , r_min , r_max = is_bst(node.right)
    is_bst_node = is_bst_l and is_bst_r and (l_max is None or node.key > l_max) and (r_min is None or r_min < node.key)
    min_key=min(l_min,node.key,r_min )
    max_key=min(l_max,node.key,r_max )
    return is_bst_node , min_key,max_key
```
