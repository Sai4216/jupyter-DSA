```python
def fib(n):
    if n == 0 :
        return 0
    elif n == 1 or n == 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)
fib(9)
```

```python
def fact(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n*fact(n-1)
```

```python
s = 'sai'
vowel = ['A','E','I','O','U','a','e','i','o','u']
c = 0
for i in s:
    if i in vowel:
        c += 1
print(fact(len(s)-c))
```

```python
n = [10,20,30,40,50,60]
n.reverse()
sum = 0
for i in range(len(n)):
    if i%2==0:
        sum += n[i]
print(sum)
```
