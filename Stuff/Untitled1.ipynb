```python
def remove_None(nums):
    return [x for x in nums if x is not None]

def is_bst(node):
    if node is None:
        return True,None,None
    is_bst_l , l_min , l_max = is_bst(node.left)
    is_bst_r , r_min , r_max = is_bst(node.right)
    is_bst_node = is_bst_l and is_bst_r and (l_max is None or node.key > l_max) and (r_min is None or node.key < r_min )
    min_key=min(remove_None([l_min,node.key,r_min] ))
    max_key=max(remove_None([l_max,node.key,r_max]) )
    return is_bst_node , min_key,max_key
```

```python
def is_balanced(node):
    if node is None:
        return True , 0
    left_tree , l_height = is_balanced(node.left)
    right_tree , r_height = is_balanced(node.right)
    balanced = left_tree and right_tree and abs(l_height-r_height) <= 1
    height = 1 + max(l_height,r_height)
    return balanced,height
```

```python
class TreeNode:
    def __init__(self,key):
        self.left = None
        self.key = key
        self.right = None
```

```python
node0 = TreeNode(3)
node1 = TreeNode(9)
node2 = TreeNode(20)
node3 = TreeNode(15)
node4 = TreeNode(7)
```

```python
node0.left = node1
node0.right = node2
node2.left = node3
node2.right = node4
```

```python
tree1 = node0
```

```python
def inOrder(node):
    if node is None:
        return []
    return (inOrder(node.left) + [node.key] + inOrder(node.right))
```

```python
inOrder(tree1)
```

```python
is_balanced(tree1)
```

```python
is_bst(tree1)
```
