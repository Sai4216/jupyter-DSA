```python
def fun(arr):
    max = 0
    i , j = 0 , 0
    curr = arr[0]
    while i<len(arr):
        c = 0
        while j < len(arr):
            if arr[j] > curr:
                curr = arr[j]
                c += 1
            j += 1
        if c > max:
            max = c
        i += 1
    return max + 1
print(fun([5,2,8,6,3,6,9,5]))
```

```python
def helper(c,wt,val,n):
    if n == 0 or c == 0:
        return 0
    if c >= wt[n-1]:
        return max(val[n-1] + helper(c-wt[n-1],wt,val,n-1),helper(c,wt,val,n-1))
    else:
        return helper(c,wt,val,n-1)
def knapSack(c, val, wt):
    return helper(c,wt,val,len(val))

c = 5
val = [10, 40, 30, 50]
wt = [5, 4, 6, 3] 
print(knapSack(c,val,wt))
```

```python
def knapsack(c,val,wt):
    table = [[0 for _ in range(c + 1)] for _ in range(len(val) + 1)]
    for i in range(len(val)):
        for j in range(1,c + 1):
            if c > wt[i]:
                table[i+1][j] = table[i][j]
            else:
                table[i+1][j] = max(table[i][j],val[i]+table[i][j-wt[i]])
    print(table)
    print(table[-1][-1])
knapsack(c,val,wt)
```

```python
def rotate(matrix):   
    row , col = len(matrix) , len(matrix[0])
    for r in range(row):
        for c in range(r,col):
            matrix[r][c] , matrix[c][r] = matrix[c][r] ,matrix[r][c]
    for i in matrix:
        i.reverse()
    
    print(matrix)
rotate([[1,2,3],[4,5,6],[7,8,9]])
```
