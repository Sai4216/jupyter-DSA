```python
from collections import defaultdict
node_count = 4
edge_count = 5
adjacency_map =  defaultdict(set)
all_nodes =  set()
nodes = [[1,3],[3,2],[0,3],[0,1],[2,1]]
for node in nodes:
node1,node2 = node
adjacency_map[node1].add(node2)
adjacency_map[node2].add(node1)
all_nodes.add(node1)
all_nodes.add(node2)
target_count= 8
active_nodes  = list(all_nodes)
processed_count = 0
days_elapsed = 0
inactive_nodes = []
node_status =  {node: 1 for node in active_nodes}
print(node_status)
while processed_count < target_count:
    
  print('hi')
    days_elapsed += 1
    next_active_nodes = []
    next_inactive_nodes = []
    for node in active_nodes:
      processed_count += 1
      print('hi')
      neighbors = adjacency_map[node]
      active_neighbor_count = sum(node_status[neighbor] == 1 for neighbor in neighbors)
      if active_neighbor_count == 3:
        next_active_nodes.append(node)
      else:
        next_inactive_nodes.append(node)
    for node in inactive_nodes:
      neighbors = adjacency_map[node]
      active_neighbor_count = sum(node_status[neighbor] == 1 for neighbor in neighbors)
      if active_neighbor_count == 3:
        next_active_nodes.append(node)
      else:
        next_inactive_nodes.append(node)
    active_nodes = next_active_nodes
    inactive_nodes = next_inactive_nodes
    for node in active_nodes:
      node_status[node] = 1
    for node in inactive_nodes:
      node_status[node] = 0
    print(days_elapsed)
```
