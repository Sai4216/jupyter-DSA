```python
from collections import defaultdict, deque
import sys

def min_cost_to_form_string(n, substrings, main_string):
    # Initialize dp table
    dp = [float('inf')] * (len(main_string) + 1)
    dp[0] = 0

    # Iterate through the main string
    for i in range(len(main_string) + 1):
        if dp[i] == float('inf'):
            continue

        # Check substrings to find matches
        for substring, cost in substrings:
            length = len(substring)
            if i + length <= len(main_string) and main_string[i:i + length] == substring:
                dp[i + length] = min(dp[i + length], dp[i] + cost)

    # If the last position in the DP table is infinity, it's impossible to form the main string
    return dp[-1] if dp[-1] != float('inf') else "Impossible"

# Input handling
n = int(input())
substrings = []
for _ in range(n):
    line = input().split()
    substrings.append((line[0], int(line[1])))
main_string = input()

# Solve the problem
result = min_cost_to_form_string(n, substrings, main_string)
print(result)
```
