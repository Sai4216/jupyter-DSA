```python

def completeChallenge(input1, input2):
    # input1: number of books (N)
    # input2: list of pages in each book (A)
    def is_possible(pages, max_pages, days):
        cnt, curr_sum = 1, 0
        for p in pages:
            if curr_sum + p > max_pages:
                cnt += 1
                curr_sum = p
            else:
                curr_sum += p
        return cnt <= days

    N = input1
    A = input2
    days = 5
    left, right = max(A), sum(A)
    result = right
    while left <= right:
        mid = (left + right) // 2
        if is_possible(A, mid, days):
            result = mid
            right = mid - 1
        else:
            left = mid + 1
    return result
completeChallenge(5,)
```

```python
def split_into_parts(input, num_parts):
    base = input // num_parts  # 21
    remainder = input % num_parts  # 0
    result = [base] * num_parts
    for i in range(remainder):
        result[i] += 1
    return result

print(split_into_parts(106, 5))  # Output: [21, 21, 21, 21, 21]
```

```python
def split_into_parts(input1, input2):
    input2 = sum(input2)
    base = input2 // input1
    remainder = input2 % input1
    result = [base] * input1
    for i in range(remainder):
        result[i] += 1
    return min(result)

print(split_into_parts(5, 105))

```

```python
def f(input2):
    n = len(input2)
    lv = 0
    
    for i in range(n):
        lv = lv^input2[i]
        rv = 0
        for j in range(i+1,n):
            rv = rv^input2[j]
        
        if lv > rv:
            return i
    return -1
    
f([0,0,0,0])
```
