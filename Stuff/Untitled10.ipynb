```python
from collections import defaultdict
import math

def calculate_voltage(n, wires, animals, target_animal):
    intersections = defaultdict(list)

    # Helper function to check if two wires intersect
    def wire_intersects(w1, w2):
        (x1, y1, x2, y2) = w1
        (x3, y3, x4, y4) = w2

        def on_segment(p, q, r):
            if (q[0] <= max(p[0], r[0]) and q[0] >= min(p[0], r[0]) and
                q[1] <= max(p[1], r[1]) and q[1] >= min(p[1], r[1])):
                return True
            return False

        def orientation(p, q, r):
            val = (q[1] - p[1]) * (r[0] - q[0]) - (q[0] - p[0]) * (r[1] - q[1])
            if val == 0: return 0
            return 1 if val > 0 else 2

        p1, q1 = (x1, y1), (x2, y2)
        p2, q2 = (x3, y3), (x4, y4)

        o1 = orientation(p1, q1, p2)
        o2 = orientation(p1, q1, q2)
        o3 = orientation(p2, q2, p1)
        o4 = orientation(p2, q2, q1)

        if o1 != o2 and o3 != o4:
            return True

        if o1 == 0 and on_segment(p1, p2, q1): return True
        if o2 == 0 and on_segment(p1, q2, q1): return True
        if o3 == 0 and on_segment(p2, p1, q2): return True
        if o4 == 0 and on_segment(p2, q1, q2): return True

        return False

    # Find intersections between wires
    for i in range(n):
        for j in range(i + 1, n):
            if wire_intersects(wires[i], wires[j]):
                intersections[(i, j)].append(wires[i])
                intersections[(i, j)].append(wires[j])

    # Calculate voltage for each intersection
    total_voltage = 0

    for key, wires_at_point in intersections.items():
        min_cells = float('inf')
        wire_count = len(wires_at_point)
        for wire in wires_at_point:
            (x1, y1, x2, y2) = wire
            cells = abs(x2 - x1) + abs(y2 - y1)
            min_cells = min(min_cells, cells)
        total_voltage += wire_count * min_cells

    # Check if the target animal dies
    target_threshold = animals[target_animal]
    animal_dies = total_voltage > target_threshold

    # Calculate probability of animals dying
    deaths = sum(1 for animal, threshold in animals.items() if total_voltage > threshold)
    probability = round(deaths / len(animals), 2)

    return ("Yes" if animal_dies else "No", probability)

# Input handling


n = int(input())
wires = []
for _ in range(n):
    wires.append(tuple(map(int, input().split())))

animal_data = input().split()
animals = {entry.split(':')[0]: int(entry.split(':')[1]) for entry in animal_data}

target_animal = input().strip()

result, probability = calculate_voltage(n, wires, animals, target_animal)
print(result,end="\n")
print(probability,end = "")
```
