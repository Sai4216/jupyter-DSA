```python
def fib(n):
    if n == 0:
        return 0
    elif n == 2 or n == 1:
        return 1
    else:
        return fib(n-1) + fib(n-2)
print(fib(10))
```

```python
def B(list):
    for i in range(len(list) - 1):
        for j in range(len(list) - 1 ):
            if list[j] > list[j+1]:
                list[j] , list[j+1] = list[j+1] , list[j]
    print(list)
```

```python
def bs(arr,key):
    l = 0
    h = len(arr) - 1
    while l <= h:
        mid = (l+h) // 2
        if key == arr[mid]:
            return mid 
        elif key>arr[mid]:
            l = mid+1
        else:
            h = mid - 1
    return -1
bs([1,5,3,4,8,6,8],6)
```
