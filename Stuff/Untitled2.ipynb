```python
def fib(n):
    if n == 0 :
        return 0
    elif n == 1 or n == 2:
        return 1
    else:
        return fib(n-1) + fib(n-2)
fib(9)
```

```python
s = 'sai'
vowel = ['A','E','I','O','U','a','e','i','o','u']
c = 0
for i in s:
    if i in vowel:
        c += 1
print(fact(len(s)-c))
```
