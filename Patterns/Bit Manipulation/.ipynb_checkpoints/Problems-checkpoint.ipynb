```python
#L50
def myPow(x,n):
    nev = False if n > 0 else True
    n = abs(n)
    ans = 1
    while n>0:
        if n&1:
            n = n-1
            ans *= x
        else:
            n = n//2
            x *= x
    return ans if not nev else 1/ans
x = 2
n = 5
myPow(x,n)
```

```python
#L78
def subsets(nums):
    n = len(nums)
    subsets = 1<<n
    ans = []
    for i in range(subsets):
        list = []
        for j in range(n):
            if i&(1<<j):
                list.append(nums[j])
        ans.append(list)
    return ans
nums = [1,2,3]
subsets(nums)
```

```python
#L136
def singleNumber(nums):
    ans = 0
    for i in nums:
        ans ^= i
    return ans
arr = [1,2,2,3,3]
singleNumber(arr)
```

```python
#L2220
def minBitFlips(start, goal):
    n = start^goal
    cnt = 0
    while n != 0:
        n = n&(n-1)
        cnt += 1
    return cnt
start = 10
goal = 7
minBitFlips(start,goal)
```

```python
#L231
def isPowerOfTwo(self, n):
    return n>0 and n&(n-1) == 0
isPowerOfTwo(16)
```
