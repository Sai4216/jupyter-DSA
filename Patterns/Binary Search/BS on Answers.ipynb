```python
#L4
def findMedianSortedArrays(nums1,nums2):
    result = []
    i,j = 0,0
    while i < len(nums1) and j <len(nums2):
        if nums1[i] < nums2[j]:
            result.append(nums1[i])
            i += 1
        else:
            result.append(nums2[j])
            j += 1
    while i<len(nums1):
        result.append(nums1[i])
        i += 1
    while j<len(nums2):
        result.append(nums2[j])
        j += 1
    n = len(result)
    if n % 2 != 0:
        return float(result[n//2]) 
    else:
        return (result[n//2] + result[n//2-1]) / 2.0
nums1 = [1,2]
nums2 = [3,4]
findMedianSortedArrays(nums1,nums2)
```

```python
#L410
def numOfSplits(nums,msum):
    c = 1
    sum = 0
    for i in nums:
        if sum + i<=msum:
            sum += i
        else:
            sum = i
            c += 1
    return c
def splitArray(nums, k):
    n = len(nums)
    low = max(nums)
    high = sum(nums)
    while low <= high:
        mid = (low+high) // 2
        if numOfSplits(nums,mid) <= k:
            high = mid-1
        else:
            low = mid+1
    return low

nums = [1,2,3,4,5]
k = 2
splitArray(nums,k)
```

```python
#L1539
def findKthPositive(arr,k):
    low = 0
    high = len(arr)
    n = len(arr)
    while low<=high:
        mid = (low+high) // 2
        if mid >= n:
            return k+n
        if arr[mid] - (mid+1) < k:
            low = mid+1
        else:
            high = mid-1
    return low+k
arr = [2,3,4,7,11]
k = 5
findKthPositive(arr,k)
```

```python
#L1011
def Possible(weights,days,capacity):
    idx = 0
    days = days-1
    temp = capacity
    while idx < len(weights):
        if weights[idx] > capacity:
            days -= 1
            capacity = temp
        capacity -= weights[idx]
        idx += 1
        if capacity < 0:
            return False
        # print("day capacity",day,capacity)
    return True if days >= 0 else False
    
def shipWithinDays(weights, days):
    low = sum(weights) // days
    high = sum(weights)
    while low <= high:
        capacity = (low+high) // 2
        # print("low {} high {} mid {}".format(low,high,capacity))
        if Possible(weights,days,capacity):
            high = capacity-1
        else:
            low = capacity+1
            
    return low
weights = [1,2,3,1,1]
days = 4
shipWithinDays(weights,days)
```

```python
# L1283
import math
def sumDivisors(nums,divisor):
    sum = 0
    for i in nums:
        sum += math.ceil(i/divisor)
    return sum
    
def smallestDivisor(nums,threshold):
    low = 1
    high = max(nums)
    while low <= high:
        mid = (low+high)//2
        if sumDivisors(nums,mid) > threshold:
            low = mid+1
        else:
            high = mid-1
    return low
nums = [21212,10101,12121]
threshold = 1000000
smallestDivisor(nums,threshold)
```

```python
# L1482
def totalBloom(bloomDay,day,k):
    c = 0
    bouquets = 0
    for i in bloomDay:
        if day//i >= 1:
            c += 1
        else:
            c = 0
        if c == k:
            bouquets += 1
            c = 0
    return bouquets

def minDays(bloomDay, m, k):
    low = min(bloomDay)
    high = max(bloomDay)
    while low<=high:
        day = (low+high) // 2
        if totalBloom(bloomDay,day,k) >= m:
            high = day-1
        else:
            low = day+1
        
    return -1 if m*k > len(bloomDay) else low

bloomDay = [62,75,98,63,47,65,51,87,22,27,73,92,76,44,13,90,100,85]
m = 2
k = 7
minDays(bloomDay,m,k)
```

```python
# L875
import math
def totalHours(piles,k):
    total = 0
    for i in piles:
        total += math.ceil(i/k)
    return total
    
def f(piles,h):
    low = 1
    high = max(piles)
    while low <= high:
        mid = (low+high) // 2
        if totalHours(piles,mid) <= h:
            high = mid-1
        else:
            low = mid+1
    return low
piles = [30,11,23,4,20]
h = 6
f(piles,h)
```
