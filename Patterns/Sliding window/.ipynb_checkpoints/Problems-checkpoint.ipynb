```python
#L643
def f(nums,k):
    max = -float("inf")
    n = len(nums)
    sum = 0
    maxSum = 0
    for i in range(k):
        sum += nums[i]
    maxSum = sum
    for i in range(k,n):
        sum += nums[i] - nums[i-k]
        if sum > maxSum:
            maxSum = sum
    return maxSum/float(k)

nums = [5]
k = 1
f(nums,k)
```

```python
#L3
from collections import deque
def f(s):
    left = 0
    l = deque()
    count = 0
    for right in range(len(s)):
        while s[right] in l:
            l.popleft()
            left += 1
        l.append(s[right])
        count = max(count,right-left+1)
        print(l)
    return count

s = "aab"
f(s)
```
