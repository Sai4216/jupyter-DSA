```python
#L131
def isPalindrome(s):
    return s == s[::-1]

def helper(ind,ds,s,ans,n):
    if ind >= n:
        ans.append(list(ds))
        return
    for i in range(ind,n):
        if isPalindrome(s[ind:i+1]):
            ds.append(s[ind:i+1])
            helper(i+1,ds,s,ans,n)
            ds.pop()
    
def partition(s):
    ans = []
    helper(0,[],s,ans,len(s))
    return ans
s = "aab"
partition(s)
```
