```python
#L39
def helper(i,target,arr,n,ds,ans):
    if i >= n:
        if target == 0:
            ans.append(list(ds))
        return
    if target >= arr[i]:
        ds.append(arr[i])
        helper(i,target - arr[i],arr,n,ds,ans)
        ds.pop()
    helper(i+1,target,arr,n,ds,ans)

def combinationSum(candidates, target):
    ans = []
    helper(0,target,candidates,len(candidates),[],ans)
    return ans
combinationSum([2,3,6,7],7)
```

```python
#L78
def helper(i,n,nums,ds,ans):
    if i>=n:
        ans.append(list(ds))
        return
    ds.append(nums[i])
    helper(i+1,n,nums,ds,ans)
    ds.pop()
    helper(i+1,n,nums,ds,ans)
    return ans
def subsets(nums):
    return helper(0,len(nums),nums,[],[])
subsets([1,2,3])
```

```python
#L22
def valid(s):
        cnt = 0
        for i in s:
            if cnt < 0:
                return False
            if i == "(":
                cnt += 1
            else:
                cnt -= 1
        return cnt == 0
def helper(i,s,n,ans):
    if i >= n:
        if valid(s):
            ans.append(s)
        return
    helper(i+1,s+"(",n,ans)
    helper(i+1,s+")",n,ans)
    return ans
def generateParenthesis(n):
    ans = []
    helper(0,"",2*n,ans)
    return ans
generateParenthesis(3)
```
